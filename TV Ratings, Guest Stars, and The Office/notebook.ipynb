```python
# Use this cell to begin your analysis, and add as many as you would like!

# Import pandas and matplotlib.pyplot
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Read in the csv file as DataFrame
df_the_office = pd.read_csv('datasets/office_episodes.csv', parse_dates=['release_date'])
df_the_office.head() 
```

```python
# Initialze two empty lists to contain my colors and sizes
cols = []
sizes = []

# Iterate through the DataFrame, and assign colors based on ratings
for ind, row in df_the_office.iterrows():
    if row.scaled_ratings < 0.25:
        cols.append('red')
    elif row.scaled_ratings < 0.50:
        cols.append('orange')
    elif row.scaled_ratings < 0.75:
        cols.append('lightgreen')
    else:
        cols.append('darkgreen')
        
# Iterate through the DataFrame, and assign a size on whether it has guests
for ind, row in df_the_office.iterrows():
    if row.has_guests == False:
        sizes.append(25)
    else:
        sizes.append(250)
        
# To be able to plott, we would need to add our new lists to the DataFrame
df_the_office['colors'] = cols
df_the_office['sizes'] = sizes

# Let's split the data into guest and non_guest DataFrames
non_guest = df_the_office[df_the_office.has_guests == False]
guest = df_the_office[df_the_office.has_guests == True]

# Now, we are ready to set the figure size and plot style for our visuals
plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')

# Create the figure
fig = plt.figure()

# Create two scatter plots with the episode number on the x axis, and the viewership on the y axis

# Let's create a normal scatter plot for regular episodes 
plt.scatter(x=non_guest.episode_number, y=non_guest.viewership_mil, \
               # Assign our color list as the colors and set marker and size
               c=non_guest.colors, s=25)

# Now, let's create another scatter plot for guest star episodes
plt.scatter(x=guest.episode_number, y=guest.viewership_mil, \
               # Assign color list as the colors and set marker and size
               c=guest.colors, marker='*', s=250)

# Create a title
plt.title('Popularity, Quality, and Guest Appearances on the Office')

# Create an x-axis label
plt.xlabel('Episode number', fontsize=18)

# Create a y-axis label
plt.ylabel('Viewership (Millions)', fontsize=18)

# Show plot
plt.show()

# Get the most popular guest star
print(df_the_office[df_the_office.viewership_mil > 20]['guest_stars'])
```
